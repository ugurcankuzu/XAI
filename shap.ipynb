```python
with open("datasetAndModelOut.pkl","rb") as f:
  lr = pickle.load(f)
  testX = pickle.load(f)
  trainX = pickle.load(f)
  testY = pickle.load(f)
  trainY = pickle.load(f)
```

```python
shapExp = shap.LinearExplainer(lr,trainX,feature_dependency="independent")
```

```python
shap_values = shapExp.shap_values(testX)
```

```python
shap.summary_plot(shap_values,testX.to_numpy(),testX.columns)
```

```python
shap.initjs()

shap.force_plot(shapExp.expected_value,shap_values[2],testX.columns)
```
